```python
fig, ax = plt.subplots()

date = '2024-12-18'
# file_name_list = ['#247_pulse_effect_on_sensor_dot_141640', '#248_pulse_effect_on_sensor_dot_143344', '#249_pulse_effect_on_sensor_dot_145048', 
#                   '#250_pulse_effect_on_sensor_dot_150751','#251_pulse_effect_on_sensor_dot_152455', '#252_pulse_effect_on_sensor_dot_154158', 
#                   '#253_pulse_effect_on_sensor_dot_155901', '#254_pulse_effect_on_sensor_dot_161604', '#255_pulse_effect_on_sensor_dot_163307',
#                   '#256_pulse_effect_on_sensor_dot_165010', '#257_pulse_effect_on_sensor_dot_170722']
file_name_list = ['#247_pulse_effect_on_sensor_dot_141640', '#248_pulse_effect_on_sensor_dot_143344', '#249_pulse_effect_on_sensor_dot_145048', 
                  '#250_pulse_effect_on_sensor_dot_150751','#251_pulse_effect_on_sensor_dot_152455', '#252_pulse_effect_on_sensor_dot_154158', 
                  '#253_pulse_effect_on_sensor_dot_155901', '#254_pulse_effect_on_sensor_dot_161604', '#255_pulse_effect_on_sensor_dot_163307',
                  '#256_pulse_effect_on_sensor_dot_165010']
# file_name_list = ['#248_pulse_effect_on_sensor_dot_143344', '#249_pulse_effect_on_sensor_dot_145048', '#250_pulse_effect_on_sensor_dot_150751',
#                   '#251_pulse_effect_on_sensor_dot_152455', '#252_pulse_effect_on_sensor_dot_154158', '#253_pulse_effect_on_sensor_dot_155901', 
#                   '#254_pulse_effect_on_sensor_dot_161604', '#255_pulse_effect_on_sensor_dot_163307','#256_pulse_effect_on_sensor_dot_165010']
amplitude_list = []

for file_name in file_name_list:

    data = np.load(f'C:/Users/Tatsuya Matsuda/Documents/CS_installations-HI_04Dec2024/data/{date}/{file_name}/arrays.npz')
    I = data['I']/2**12 
    Q = data['Q']/2**12 
    amplitude = np.sqrt(I**2 + Q**2)
    amplitude = amplitude[62]
    amplitude_list.append(amplitude)

time_list = [0.51, 10.5, 20.5, 30.5, 40.5, 50.5, 60.5, 70.5, 80.5, 90.5] #ms
# time_list = [10.5, 20.5, 30.5, 40.5, 50.5, 60.5, 70.5, 80.5, 90.5] #ms
# time_fit = np.linspace(10.5, 90.5, 100)
time_fit = np.linspace(0.5, 90.5, 10000)

ax.plot(time_list, amplitude_list, '-')
# ax.set_title(f'{file_name}')
ax.set_xlabel('Measurement delay (ms)')
ax.set_ylabel('Amplitude (V)')



## fitting
from lmfit import Model
def exp_model(x, a, C):
    return np.exp(-a * x) + C

# lmfitのModelオブジェクト
model = Model(exp_model)

# 初期パラメータ
params = model.make_params(a=0.1, C=0.5)  # 初期値を指定

# フィッティング
result = model.fit(amplitude_list, params, x=time_list)

# フィッティング結果
print(result.fit_report())

# フィッティング曲線
y_fit = result.best_fit

a = 34.0530947 # need to change
C = 3.6378e-05
y_fit_ac = np.exp(-a * time_fit) + C

# プロット
plt.figure()
plt.scatter(time_list, amplitude_list, label="Data", s=10, color="blue")
plt.plot(time_fit, y_fit_ac, label=f"Fit: {result.best_values}", color="red")

plt.xlabel('Measurement delay (ms)')
plt.ylabel('Amplitude (V)')
plt.legend()
plt.show()
```
